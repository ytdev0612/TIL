```python
# Pclass 시각화
```

```python
df.Pclass.value_counts()
```

```python
plt.bar(df.Pclass.unique(), df.groupby('Pclass')['Survived'].count())
plt.xticks(df.Pclass.unique(), ['1class', '2class', '3class'])
```

```python
# 데이터 생성 및 정렬
categories = df.Pclass.unique()
categories.sort()
values = df.groupby('Pclass')['Survived'].count()[categories]

# 바 차트 그리기
plt.bar(categories, values)

# x축 눈금 레이블 설정
plt.xticks(categories, ['1class', '2class', '3class'])

# 그래프 보여주기
plt.show()
```

```python
# Sex 시각화
```

```python
categories = df.Sex.unique()
values = df.Sex.value_counts()
```

```python
categories = df.Sex.unique()
values = df.Sex[df.Survived==1].value_counts()
```

```python
categories = df.Sex.unique()
values = df.Sex[df.Survived==0].value_counts()
```

```python
plt.bar(categories, values)
```

```python
categories = ['male_alive', 'male_dead', 'female_alive', 'female_dead']
values = [df[(df.Sex=='male') & (df.Survived==1)].shape[0],
          df[(df.Sex=='male') & (df.Survived==0)].shape[0],
          df[(df.Sex=='female') & (df.Survived==1)].shape[0],
          df[(df.Sex=='female') & (df.Survived==0)].shape[0]]

# 바 차트 그리기
plt.bar(categories, values)

# 그래프 보여주기
plt.show()
```
