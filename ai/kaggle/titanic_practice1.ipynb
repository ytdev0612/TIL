```python
# 결측값 채우기
train_df[(train_df.Pclass==1) & (train_df.Age.isnull())]
```

```python
# 성별과 티켓 등급별 나이 평균 계산
mean_ages = train_df.groupby(['Sex', 'Pclass'])['Age'].transform('mean')

# NaN 값에 평균 나이 채워 넣기
train_df['Age'].fillna(mean_ages)
```

```python
# 성별에 따라 그룹화하고 요금에 따라 구간 나누기
fare_bins = train_df.groupby('Sex')['Fare'].transform(lambda x: pd.cut(x, 5))

# 각 구간의 나이 평균 계산
mean_ages = train_df.groupby(fare_bins)['Age'].transform('mean')

# NaN 값에 평균 나이 채워 넣기
train_df['Age'].fillna(mean_ages, inplace=True)
```

```python
# 결측값 채우기
# 성별에 따라 그룹화하고 요금에 따라 구간 나누기
fare_bins = test_df.groupby('Sex')['Fare'].transform(lambda x: pd.cut(x, 5))

# 각 구간의 나이 평균 계산
mean_ages = test_df.groupby(fare_bins)['Age'].transform('mean')

# NaN 값에 평균 나이 채워 넣기
test_df['Age'].fillna(mean_ages,  inplace=True)
```

```python
# Pclass가 3인 여성 중에서 완전히 혼자온사람중 항구가 C,Q면 생존확률이 높았다
# p3_fe[(p3_fe.SibSp == 0) & (p3_fe.Parch == 0)].groupby('Embarked').Survived.value_counts()
a = test_df[(test_df.Pclass==3) & (test_df.Sex=='female')]
a = a[(a.SibSp == 0) & (a.Parch == 0)]
alive_id2 = a[(a.Embarked == 'C') | (a.Embarked == 'Q')].PassengerId
```

```python
gender_submission.loc[gender_submission['PassengerId'].isin(alive_id2), 'Survived'] = 1
```

```python
gender_submission.to_csv('my_submission.csv', index=False)
# 0.77 소폭하락, 취소 (탑승항구라서 생존률과는 의미없는 결과?)
```

```python
# Pclass1,2인 남성 18세이하 아이는 생존확률이 높다
male1 =  test_df[((test_df.Pclass==1)|(test_df.Pclass==2)) & (test_df.Sex=='male')]
alive_id3 = male1[male1.Age<18].PassengerId
```
