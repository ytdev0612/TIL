```python
import pandas as pd
import numpy as np
```

```python
rating_data = pd.read_csv('Movielens/ratings.csv')
rating_data # 평가 데이터
```

```python
movie_data = pd.read_csv('Movielens/movies.csv')
movie_data # 영화 데이터
```

```python
# 필요없는 열 제거
rating_data.drop('timestamp', axis=1, inplace=True)
```

```python
# movie_user_rating.head()
# Leaving Las Vegas (1995) 에 대한 평가 점수 분포와 가장 비슷한 평가 점수 분포를 띄는 영화를 검색하여 추천
movie_user_rating.loc['Leaving Las Vegas (1995)']
# 유클리디안 거리, 코사인 유사도 
# 영화행 영화열 -> 코사인 유사도 행렬
```

```python
# 예시
#         딥러닝   머신러닝  나는
# 영화1     0          1       1
# 영화2     1          0       1 
# 영화3     2          2       2  
# 영화4     1          1       1
# 3차원(딥러닝축, 머신러닝축, 나는축) 벡터 공간 -> 벡터가 4개, 벡터간의 거리

# 영화4와 가장 거리(유클리디안)가 가까운(유사한) 영화가 무엇인가요?
# 영화1,2(거리 : 1)

# 코사인 유사도로 하면 3과4가 가까움(실제는 크기도 봄)
from numpy.linalg import norm

def cossim(a,b):
    return np.dot(a,b) / (norm(a) * norm(b))

m1=np.array([0,1,1])
m2=np.array([1,0,1])
m3=np.array([2,2,2])
m4=np.array([1,1,1])
print(cossim(m1,m4))
print(cossim(m2,m4))
print(cossim(m3,m4))
```

```python
from sklearn.metrics.pairwise import cosine_similarity # 코사인 유사도 구해주는 함수
```

```python
item_based_cola_sim = cosine_similarity(movie_user_rating)
item_based_cola_sim
```

```python
item_based_cola_sim.shape
```

```python
item_based_cola_sim[0]
```

```python
movie_user_rating.loc['Leaving Las Vegas (1995)']
# Leaving Las Vegas 영화에 대한 코사인 유사도가 가장 높은 한 편의 영화 제목을 출력하시오(자신 제외)
```

```python
movie_user_rating.index[item_based_cola_sim[movie_user_rating.index.get_loc('Leaving Las Vegas (1995)')].argsort()[-2]]
```

```python
cosine_df = pd.DataFrame(item_based_cola_sim, index=movie_user_rating.index, columns=movie_user_rating.index)
cosine_df.drop('Leaving Las Vegas (1995)', axis=1, inplace=True)
# 코사인 유사도가 가장 큰 인덱스의 영화를 추출
cosine_df.loc['Leaving Las Vegas (1995)'].sort_values(ascending=False).index[0]
```
