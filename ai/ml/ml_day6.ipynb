```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('GroceryStoreDataSet.csv', names=['products'], sep=',')
df
```

```python
data = list(df['products'].apply(lambda x: x.split(','))) # 이중리스트로 df['products']값이 들어감
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
trans_data = te.fit(data).transform(data)
trans_data # 총 아이템 종류 : 11개, 첫번째 array에 True가 Milk, Bread, Biscuit
```

```python
te.columns_ # 아이템의 총 종류(알파벳 순)
```

```python
df = pd.DataFrame(trans_data, columns=te.columns_)
df # 여기까지 준비과정
```

```python
# transaction dataframe 이전까지 한 내용
# 비 빈번 항목 날리기
# 1 : apriori -> 지지도 설정(0.2)
# 2 : association rules -> 신뢰도 설정(0.6)
# 3 : lift -> 향상도 설정(2)
# 규칙 추론
```

```python
df = apriori(df, min_support=0.2, use_colnames=True)
df # 최소 지지도가 0.2 이상인 아이템셋 출력
```

```python
df_ar = association_rules(df, metric='confidence', min_threshold=0.6)
df_ar # 공식으로 생각했을 때 an~, co~ 가 x,y, 각각 support, 동시에 support, 신뢰도(0.6이상, 0.8이면 우유를 사면 0.8빵을 삼),
      # lift(우유를 구매한 사람중 빵을 구매한 사람/전체 고객대비 빵구매)
```

```python
# 주문번호가 같으면 하나의 주문
# 아이템의 개수(개수는 무시)
# choice_description [음식이름, [아마 구성성분(재료)?]]
# 매출액을 올리고 싶다 -> 매출액을 올리기 위한 연관 규칙 찾기(ex: 어떤걸 사면 어떤걸 같이 삿다, choice_description에 내용물 추천, ...)
```
